Objective    : To use the Seaborn library to find patterns in the Titanic<br>

```python
# Explanation And Illustration:
# In this Experiment we use the inbuilt Titanic dataset from Seaborn.
# The dataset contains 891 rows with information about passengers
# who boarded the Titanic ship. We use Seaborn and Matplotlib
# to visualize patterns in the data and plot a histogram
# for the distribution of ticket fare (fare column).
```

```python
# Steps: 1) Import Libraries (seaborn, matplotlib, pandas, numpy)
#        2) Load the Titanic dataset using seaborn
#        3) Display first 5 rows using df.head()
#        4) Check shape and basic info of dataset
#        5) Pattern 1 - Count of Survivors vs Non-Survivors (countplot)
#        6) Pattern 2 - Survival based on Passenger Class (countplot with hue)
#        7) Pattern 3 - Survival based on Gender (countplot with hue)
#        8) Pattern 4 - Age distribution of passengers (histplot)
#        9) Histogram - Distribution of Fare (hist with matplotlib)
```

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print("Libraries Imported Successfully!")
```

```python
# Step 2: Load the Titanic Dataset
df = sns.load_dataset('titanic')
print("Dataset Loaded Successfully!")
print("Source: Seaborn inbuilt Titanic dataset")
```

```python
# Step 3: View first 5 rows
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# Step 4: Check shape and basic info
print("\nShape of Dataset (rows, columns):")
print(df.shape)

print("\nColumn Names and Data Types:")
print(df.dtypes)

print("\nMissing Values in each column:")
print(df.isnull().sum())
```

```python
# Step 5: Pattern 1 - Count of Survivors vs Non-Survivors
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df, palette='Set2')
plt.title('Count of Survivors vs Non-Survivors')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
print("Observation: More passengers did NOT survive than survived.")
```

```python
# Step 6: Pattern 2 - Survival based on Passenger Class
plt.figure(figsize=(7, 4))
sns.countplot(x='pclass', hue='survived', data=df, palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class (1 = First, 2 = Second, 3 = Third)')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
print("Observation: First class passengers had a higher survival rate.")
print("Third class passengers had the highest number of deaths.")
```

```python
# Step 7: Pattern 3 - Survival based on Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=df, palette='Set2')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
print("Observation: Females had a much higher survival rate than males.")
print("This reflects the 'women and children first' evacuation policy.")
```

```python
# Step 8: Pattern 4 - Age distribution of passengers
plt.figure(figsize=(8, 4))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='steelblue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
print("Observation: Most passengers were between 20 and 40 years of age.")
```

```python
# Step 9: Histogram - Distribution of Fare
plt.figure(figsize=(8, 5))
plt.hist(df['fare'], bins=40, color='coral', edgecolor='black')
plt.title('Distribution of Ticket Fare for Titanic Passengers')
plt.xlabel('Fare (Ticket Price)')
plt.ylabel('Number of Passengers')
plt.tight_layout()
plt.show()

print("\nFare Statistics:")
print(f"  Minimum Fare  : {df['fare'].min():.2f}")
print(f"  Maximum Fare  : {df['fare'].max():.2f}")
print(f"  Mean Fare     : {df['fare'].mean():.2f}")
print(f"  Median Fare   : {df['fare'].median():.2f}")
print("\nObservation: Most passengers paid a low fare (below 50).")
print("A small number of passengers paid very high fares (outliers above 300).")
print("The fare distribution is right-skewed (positively skewed).")
```
