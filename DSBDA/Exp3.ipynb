Experiment 3 : Descriptive Statistics<br>
DataSet      : titanic and iris Datasets (from sklearn)<br>
Objective    : To Measure the Central Tendencies of the dataset

```python
#Explaination And Illustration :
# In this Experiment We are just calculating the Central Tendencies like
# mean, median, minimum, maximum, standard deviation and percentile, mean,
# standard deviation  for the Iris dataset
```

```python
# Steps :1) PART A :Summary Statistics gruped by male and female for the titanic dataset
#        2) PART B :Iris dataset Statistics
```

```python
# Step 1:Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# PART A: Summary statistics grouped by categorical variable
# Load Titanic dataset
df = sns.load_dataset('titanic')
print("Part 1: Summary Statistics Grouped by Category\n")

# Fill missing age values
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
# 1) Group by 'sex' and get summary stats of 'fare'
grouped = df.groupby('sex')['fare'].agg([
    ('Mean', 'mean'),
    ('Median', 'median'),
    ('Minimum', 'min'),
    ('Maximum', 'max'),
    ('Std Deviation', 'std')
])
print("Summary Statistics of Fare grouped by Sex:")
print(grouped)
```

```python
# 2) Group by 'pclass' and get summary stats of 'age'
grouped2 = df.groupby('pclass')['age'].agg([
    ('Mean', 'mean'),
    ('Median', 'median'),
    ('Minimum', 'min'),
    ('Maximum', 'max'),
    ('Std Deviation', 'std')
])
print("\nSummary Statistics of Age grouped by Passenger Class:")
print(grouped2)
```

```python
# Create a list of numeric values for each category
print("\nList of fare values by sex:")
for sex, group in df.groupby('sex'):
    print(f"{sex}: {list(group['fare'].head(5))}")  # showing first 5
```

```python
# PART 2: Iris dataset statistics
print("\n Part 2: Iris Dataset Statistics\n")

iris = sns.load_dataset('iris')
print("Iris Dataset - First 5 rows:")
print(iris.head())

# Statistics for each species
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    print(f"\n--- Species: {species} ---")
    print(f"Mean:\n{subset.describe().loc['mean']}")
    print(f"\nStd Dev:\n{subset.describe().loc['std']}")
    print(f"\nPercentiles (25%, 50%, 75%):")
    print(subset.describe().loc[['25%', '50%', '75%']])

# Full description per species
print("\nComplete Statistics for all species:")
print(iris.groupby('species').describe())
```
