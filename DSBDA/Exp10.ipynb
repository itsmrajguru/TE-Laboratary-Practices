```python
# Steps :
# 1) Import Libraries
#         where pandas    --> Handle the data
#               numpy     --> Calculations
#               seaborn   --> Dataset and visualization
#               matplotlib --> Plotting graphs
# 2) Load the Iris dataset
# 3) View first 5 rows using df.head()
# 4) Check missing values and shape
#         df.isnull().sum() --> missing values
#         df.shape          --> (rows, columns)
# 5) List features and their types
#         df.dtypes --> data types of each column
# 6) Plot histograms for each feature
#         df.hist()         --> all features at once
#         sns.histplot()    --> individual with KDE
# 7) Plot box plots for each feature
#         sns.boxplot()     --> per feature and grouped by species
# 8) Compare distributions and identify outliers
#         df.describe()     --> statistical summary
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Loading Iris dataset
df = sns.load_dataset('iris')
print("Dataset Loaded Successfully!")
print("Source: Seaborn inbuilt Iris dataset")
```

```python
# Step 3: View first 5 rows of the dataset
print("\nFirst 5 rows of data:")
print(df.head())
```

```python
# Step 4: Check missing values and shape
print("Missing values in each column:")
print(df.isnull().sum())
print("\nShape of Dataset (rows, columns):")
print(df.shape)
```

```python
# Step 5: List features and their types
print("\nData Types of each column:")
print(df.dtypes)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
nominal_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("\nNumeric Features:", numeric_cols)
print("Nominal Features:", nominal_cols)
```

```python
# Step 6: Plot histogram for all features at once
df.hist(figsize=(10, 8), bins=15, color='steelblue', edgecolor='black')
plt.suptitle('Histogram for Each Feature - Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
# Step 7A: Box plots for all features in a single figure
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols):
    plt.subplot(1, 4, i + 1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(col)
plt.suptitle('Box Plot for Each Feature - Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
# Step 7: Box plots for all features in a single figure
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols):
    plt.subplot(1, 4, i + 1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(col)
plt.suptitle('Box Plot for Each Feature - Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
# Step 8: Compare distributions and identify outliers
print("Summary Statistics:")
print(df.describe())
```

```python
# Step 9: Observations and Inferences
print("\nObservations:")
print("1. sepal_width has visible outliers in the box plot.")
print("2. petal_length and petal_width show the best separation across species.")
print("3. Iris setosa has significantly smaller petal values compared to other species.")
print("4. Iris versicolor and virginica have overlapping sepal measurements.")
print("5. petal_length distribution is bimodal - visible in histogram.")
```
