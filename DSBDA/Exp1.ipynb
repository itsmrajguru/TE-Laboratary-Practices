```python
# Step 1:Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2 : Loading titanic dataset
df = sns.load_dataset('titanic')
print("Dataset Loaded Successfully!")
print("Source: Seaborn inbuilt Titanic dataset")
```

```python
# STEP 3: Load into DataFrame and view it
print("\nFirst 5 rows of data:")
print(df.head())
```

```python
# Step 4:Data Preprocessing
# 4A) Check the Missing values using Null() function
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# 4B) Check Statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# 4C)Check dimensions
print("\nShape of Dataset (rows, columns):")
print(df.shape)
```

```python
# 4D)Variable descriptions
print("\nData Types of each column:")
print(df.dtypes)
```

```python
# STEP 5: Data Formatting & Normalization
# 5A) Counts data types number for total entries
print("\nSummary of variable types:")
print(df.dtypes.value_counts())
```

```python
# 5B) Convert 'age' column to integer (fill missing first)
df['age'] = df['age'].fillna(df['age'].median())
df['age'] = df['age'].astype(int)
print("\nAge column converted to integer.")
```

```python
# 5C) Normalize 'fare' column between 0 and 1
df['fare_normalized'] = (df['fare'] - df['fare'].min()) / (df['fare'].max() - df['fare'].min())
print("\nFare column normalized:")
print(df[['fare', 'fare_normalized']].head())
```

```python
# STEP 6: Turn Categorical Variables into Quantitative
# 'sex' column: male=1, female=0
df['sex_encoded'] = df['sex'].map({'male': 1, 'female': 0})
```

```python
# 'embarked' column: encode using get_dummies
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
```

```python
print("\nAfter encoding categorical variables:")
print(df.head())
```
