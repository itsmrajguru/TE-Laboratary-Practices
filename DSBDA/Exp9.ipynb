```python
# Explanation And Illustration :
# Data Visualization II: Using box plots to understand distribution of age
# with respect to gender and survival status from the Titanic dataset
```

```python
# Steps :
# 1) Import Libraries
#         where pandas   --> Handle the data
#               numpy    --> Calculations
#               seaborn  --> Dataset and visualization
#               matplotlib --> Plotting graphs
# 2) Load the dataset
# 3) View first 5 rows using df.head()
# 4) Check missing values using df.isnull().sum()
# 5) Plot Box Plot for age vs sex with hue as survived
#         sns.boxplot(x='sex', y='age', hue='survived', data=df)
# 6) Write observations from the plot
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Loading titanic dataset
df = sns.load_dataset('titanic')
print("Dataset Loaded Successfully!")
print("Source: Seaborn inbuilt Titanic dataset")
```

```python
# Step 3: View first 5 rows of the dataset
print("\nFirst 5 rows of data:")
print(df.head())
```

```python
# Step 4: Check missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Step 5: Handle missing values in 'age' column before plotting
df['age'] = df['age'].fillna(df['age'].median())
print("Missing age values filled with median.")
print("Missing values after filling:")
print(df[['age', 'sex', 'survived']].isnull().sum())
```

```python
# Step 6: Plot Box Plot - Age distribution by Gender and Survival
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=df, palette='Set2')
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['Not Survived (0)', 'Survived (1)'])
plt.tight_layout()
plt.show()
```

```python
# Step 7: Additional Box Plot - Using FacetGrid for better view
g = sns.FacetGrid(df, col='survived', height=5)
g.map(sns.boxplot, 'sex', 'age', palette='Set1')
g.set_axis_labels('Gender', 'Age')
g.set_titles(col_template='Survived = {col_name}')
plt.tight_layout()
plt.show()
```

```python
# Step 8: Observations from the box plot
print("\nObservations:")
print("1. Among females, survivors tend to be younger than non-survivors.")
print("2. Among males, survivors tend to be slightly younger than non-survivors.")
print("3. Female survival rate was higher compared to males across all age groups.")
print("4. Older male passengers had a lower chance of survival.")
print("5. The median age of survived females is lower than non-survived females.")
```

```python
# Step 9: Summary statistics grouped by sex and survived
print("\nSummary Statistics - Age grouped by Sex and Survived:")
print(df.groupby(['sex', 'survived'])['age'].describe())
```
