Objective    : To preprocess text data using Tokenization,<br>
                POS Tagging, Stop Words Removal, Stemming,<br>
                Lemmatization and represent documents using<br>
                TF-IDF (Term Frequency - Inverse Document Frequency)

```python
# Explanation And Illustration:
# In this Experiment we take a sample text document and apply
# various NLP preprocessing techniques to clean and prepare
# the text data. Finally we convert text into numeric form
# using TF-IDF so that machines can understand it.
```

```python
# Steps: 1) Import Libraries (nltk, sklearn, pandas)
#        2) Define Sample Document (raw text)
#        3) Tokenization
#           - Sentence Tokenization: split text into sentences
#           - Word Tokenization: split text into individual words
#        4) POS Tagging
#           - Label each word with its grammatical role
#           - NN=Noun, VBZ=Verb, JJ=Adjective etc.
#        5) Stop Words Removal
#           - Remove common meaningless words like is, the, a
#        6) Stemming
#           - Cut words to root form (learning → learn)
#           - Uses PorterStemmer
#        7) Lemmatization
#           - Convert to proper dictionary base form (better → good)
#           - Uses WordNetLemmatizer
#        8) TF-IDF Representation
#           - TF = word frequency in document
#           - IDF = rarity of word across all documents
#           - TF-IDF = TF x IDF (higher = more important)
```

```python
import nltk
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download required NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
# Sample Document
text = """Machine learning is a subset of artificial intelligence. 
It allows computers to learn from data without being explicitly programmed. 
Deep learning is a subset of machine learning that uses neural networks."""

print("=== ORIGINAL TEXT ===")
print(text)
```

```python
# STEP 1: Tokenization
print("\n Step 1: Tokenization---")

# Sentence Tokenization
sent_tokens = sent_tokenize(text)
print("Sentence Tokens:")
for i, sent in enumerate(sent_tokens):
    print(f"  Sentence {i+1}: {sent}")

# Word Tokenization
word_tokens = word_tokenize(text)
print(f"\nWord Tokens ({len(word_tokens)} words):")
print(word_tokens)
```

```python
# STEP 2: POS Tagging
print("\n=== STEP 2: POS TAGGING ===")
pos_tags = pos_tag(word_tokens)
print("POS Tags:")
print(pos_tags)
print("\nPOS Tag Meanings:")
print("NN=Noun, VBZ=Verb, JJ=Adjective, IN=Preposition, DT=Determiner")
```

```python
# STEP 3: Stop Words Removal
print("\n=== STEP 3: STOP WORDS REMOVAL ===")
stop_words = set(stopwords.words('english'))
print(f"Total Stop Words in English: {len(stop_words)}")
print(f"Examples: {list(stop_words)[:10]}")

filtered_words = [word for word in word_tokens
                  if word.lower() not in stop_words
                  and word.isalpha()]
print(f"\nBefore removal: {word_tokens}")
print(f"\nAfter removal: {filtered_words}")
```

```python
# STEP 4: Stemming
print("\n=== STEP 4: STEMMING ===")
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("Stemmed Words:")
for original, stemmed in zip(filtered_words, stemmed_words):
    print(f"  {original} --> {stemmed}")
```

```python
# STEP 5: Lemmatization
print("\n=== STEP 5: LEMMATIZATION ===")
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("Lemmatized Words:")
for original, lemma in zip(filtered_words, lemmatized_words):
    print(f"  {original} --> {lemma}")

# Comparison: Stemming vs Lemmatization
print("\nComparison - Stemming vs Lemmatization:")
comparison = pd.DataFrame({
    'Original': filtered_words,
    'Stemmed': stemmed_words,
    'Lemmatized': lemmatized_words
})
print(comparison)
```

```python
# STEP 6: TF-IDF
print("\n=== STEP 6: TF-IDF ===")

documents = [
    "Machine learning is a subset of artificial intelligence",
    "Deep learning uses neural networks for complex tasks",
    "Artificial intelligence allows computers to learn from data",
    "Neural networks are inspired by the human brain"
]

print("Documents:")
for i, doc in enumerate(documents):
    print(f"  Doc {i+1}: {doc}")

# Calculate TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

# Display TF-IDF matrix
feature_names = tfidf.get_feature_names_out()
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names,
    index=[f'Doc{i+1}' for i in range(len(documents))]
)

print("\nTF-IDF Matrix:")
print(tfidf_df.round(4))
```
